Agregando una columna la cual nos diga cuantos generos hay en cada pelicula que el realizo

```python
def count_genres(text_):
    try:
        len_genres=len(text_.split(sep='|'))
    except:
        len_genres = np.nan
    return len_genres
```

```python
len('Hola|Mundo|Soy|Alejandro|Zavala'.split(sep='|'))
```

```python
text_test = 'Hola|Mundo|Soy|Alejandro|Zavala' # 5 is the len
count_genres(text_test)
```

```python
data_Zack["count_genres"] = data_Zack["genres"].apply(lambda x: count_genres(x))
```

```python
def every_genre(text_):
    try:
        all_genres = text_.split(sep='|')
    except:
        all_genres = []
    return all_genres
```

```python
every_genre('Hola|Mundo|Soy|Alejandro|Zavala')
```

```python
all_genre_Zack = []
for element_genre in data_Zack['genres'].tolist():
    all_genre_Zack = all_genre_Zack + every_genre(element_genre)
```
