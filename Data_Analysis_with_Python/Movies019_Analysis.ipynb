Veamos si el dataframe contiene elementos repetidos (renglones que son observacioones, registros) y eliminando registros repetidos. Ocuparemos agregaciones para saber si una pelicula aparece mas de una vez (suponiendo escritor y nombre deban ser únicas)

```python
agg_duplicated = data.groupby(["director_name", "movie_title"]).size().reset_index(name='count')
```

## Manipulando el dataframe

```python
data_zack = data[data['director_name'] == 'Zack Snyder'][['director_name','duration','movie_title',
                                                          'genres','language', 'country','content_rating',
                                                          'budget', 'title_year',
                                                          'imdb_score']]
```

```python
data_zack.head(n=25)
```

```python
data_zack = data_zack.drop_duplicates().reset_index(drop=True)
```

Ahora si, el nuevo dataframe se encuentra listo para su uso

```python
data_zack.head(n=25)
```

```python
def count_genres(str_text):
    try:
        len_genres = len(str_text.split(sep='|'))
    except:
        len_genres = np.nan
    return len_genres
```

```python
len('Hola|Mundo|Soy|Alejandro|Zavala'.split(sep='|'))
```

```python
data_zack["count_genres"] = data_zack["genres"].apply(lambda x: count_genres(x))
```

```python
data_zack
```

Encontremos todos los géneros únicos de peliculas que ha hecho Zack Snyder

```python
def text_split(str_text):
    try:
        all_genres = str_text.split(sep='|')
    except:
        all_genres = []
    return all_genres
```

```python
text_split('Hola|Mundo|Soy|Alejandro|Zavala')
```

```python
all_genre_zack = []
for element_genre in data_zack['genres'].tolist():
    all_genre_zack = all_genre_zack + text_split(element_genre)
```
