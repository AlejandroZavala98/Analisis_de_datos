```python
import os
import pandas as pd
from pandas.plotting import parallel_coordinates
from pandas_profiling import ProfileReport
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import seaborn as sns
```

```python
os.listdir()
```

```python
df_temporal = pd.read_csv(r"Insumos\temporal.csv")
display(df_temporal.head(10))
```

```python
plt.figure(figsize=(15,7))
plt.plot(df_temporal["Mes"], df_temporal["data science"],color = 'darkred')
plt.title("Serie de tiempo para data science",fontsize=20)
plt.xlabel("Tiempo", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(df_temporal["Mes"], df_temporal["data science"],color = 'darkred',label="Data Science")
plt.plot(df_temporal["Mes"], df_temporal["machine learning"],color = 'darkgreen',label="Machine Learning")
plt.plot(df_temporal["Mes"], df_temporal["deep learning"],color = 'darkcyan',label="Deep learning")
plt.title("Popularidad de terminos de Inteligencia Artificial",fontsize=20)
plt.xlabel("Tiempo", fontsize=15)
plt.ylabel("Popularidad", fontsize=15)
plt.legend(fontsize=15,title = "Categorias",title_fontsize=17)
plt.show()
```

```python
df_test = pd.DataFrame({"x":range(1,100),
                        "y":np.random.rand(99)*27 + range(1,100)})
```

```python
df_test.head()
```

```python
fig1= plt.figure(figsize=(15,7))

ax1 = fig1.add_subplot()
ax1.plot(df_test["x"],df_test["y"],"o",color="crimson")

ax1.add_patch(patches.Rectangle((20,25), #(x,y) coordenadas de inicio
                                50,# x + 50
                                100,# y + 100
                                alpha=0.35,#intensidad
                                facecolor="green",
                                edgecolor="blue",
                                linewidth=3,
                                linestyle="solid"))
ax1.set_ylabel("Muestras de un rango", fontsize=15)
ax1.set_xlabel("Observaciones de un rango", fontsize=15)
```

```python
fig2,axes2 = plt.subplots(2,2,figsize=(16,10))
fig2.suptitle('Categorias del dataframe principal',fontsize=15)
axes2[0,0].hist(df_temporal["data science"])
axes2[0,0].set_title('Histograma para data science',fontsize=13,color="darkgreen")

axes2[0,1].scatter(df_temporal["Mes"],df_temporal["data science"],color="purple")
axes2[0,1].set_title('Dispersin para data science',fontsize=13)

axes2[1,0].plot(df_temporal["Mes"],df_temporal["machine learning"],color="red")
axes2[1,0].set_title('Popularidad para machine learning',fontsize=13)

axes2[1,1].plot(df_temporal["Mes"],df_temporal["deep learning"],color="orange")
axes2[1,1].set_title('Popularidad para deep learning',fontsize=13)
```

```python
df_pokemon = pd.read_csv(r"Insumos\Pokemon.csv")
display(df_pokemon.tail(10))
```

```python
plt.figure(figsize=(15,7))
plt.scatter(df_pokemon["Attack"], df_pokemon["Speed"],color = 'darkred')
plt.title("Grfico de dispersin para el dataset pokemn",fontsize=20)
plt.xlabel("Ataque", fontsize=15)
plt.ylabel("Velocidad", fontsize=15)
plt.show()
```
