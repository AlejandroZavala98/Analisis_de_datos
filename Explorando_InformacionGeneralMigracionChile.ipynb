# Filtrando, ordenando y obteniendo muestras de un dataframe

Notemos que si nos interesara tomar una muestra de cierta cantidad de datos podemos hacer uso de la funcion sample de pandas

```python
n_total = len(data_visas)
n_total
```

```python
Queremos tomar una muestra de 10000 personas sin remmplazo
```

```python
proporcion_10000 = 10000/n_total
data_vidas_muestra = data_visas.sample(frac=proporcion_10000, replace=False)
display(data_vidas_muestra.head(15))
```

Si queremos obtener la cantidad de datos unicos de la columna (Actividad) y ademas que cuente esos valores unicos, para la muestra que obtuvimos

```python
print("Nombre de columna: {} \t Cantidad de datos: {}".format("ACTIVIDAD",data_vidas_muestra["ACTIVIDAD"].nunique()))
```

```python
data_vidas_muestra["ACTIVIDAD"].unique().tolist()
```

Viendo por posicion algunos de los paises unicos dentro de nuestra muestra

```python
data_vidas_muestra["PAIS"].unique().tolist()[15:20]
```

Tambien podemos contar los valores unicos para la columna (ACTIVIDAD) es decir cuantos de ellos son EMPLEADOS,ESTUDIANTES, ETC.

```python
data_vidas_muestra["ACTIVIDAD"].value_counts()
```

Queremos mostrar de nuestro dataframe original o filtrar por pais "MEXICO"

```python
data_visas[data_visas["PAIS"] == "MEXICO"].head(15)
```

Podemos filtrar ademas por mas de un valor, y si la linea de codigo es larga usar "comodines"

```python
tipo_actividad =['EMPLEADO',
                 'DUEÑA DE CASA',
                 'ESTUDIANTE',
                 'INACTIVO',
                 'EMPLEADO DOMESTICO',
                 'OBRERO O JORNALERO']

paises_latinoamerica = ["COLOMBIA",
                        "ARGENTINA",
                        "MEXICO",
                        "GRECIA",
                        "BRASIL"]
data_latinoamerica = data_visas[(data_visas["ACTIVIDAD"].isin(tipo_actividad)) & \
                                (data_visas["PAIS"].isin(paises_latinoamerica))]
display(data_latinoamerica.head(10))
```

```python
print("Cantidad de datos", len(data_latinoamerica))
```

```python
data_latinoamerica["ACTIVIDAD"].unique().tolist()
```

```python
data_latinoamerica["PAIS"].unique().tolist()
```
