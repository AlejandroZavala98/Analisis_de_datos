# Exploración de datos con pandas

Vamos a leer un csv directamente desde una URL que contiene información geográfica básica de los países del mundo y realizaremos un analisis de estos datos, usando la liberia pandas y matplotlib principalmente

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/lorey/list-of-countries/master/csv/countries.csv'
data = pd.read_csv(url, sep=";")
display(data.head())
```

Tambien podemos ver los datos finales de este dataframe

```python
display(data.tail())
```

Podemos ordenar un dataframe por paises de menor a mayor area

```python
by_area = data.sort_values('area')
display(by_area.head())
```

Tambien podemos ordenar de mayor a menor, por la variable "population"

```python
by_population = data.sort_values('population',ascending=False)
display(by_population.head())
```

Incluso podemos ordenar por mas de dos variables o nombres de columna

```python
by_area_population = data.sort_values(['area', 'population'])
display(by_area_population.head())
```

Veamos el numero de columnas y renglones que contiene este dataframe

```python
print('Cantidad de Filas y columnas:',data.shape)
```

Veamos el nombre de las columnas pertenecientes a este dataframe

```python
print('Nombre columnas de data:')
data.columns.tolist()
```

Viendo una previa informacion de las información por columna contenida en el dataframe, como nombre de columna, si tiene datos nulos y el tipo de variable (caracter, float, etc)

```python
data.info()
```

Pandas proporciona funciones de resumen estadisticas como por ejemplo

```python
data.describe()
```

Notemos que .describe() sirve para valores numericos (enteros y flotantes)

Podemos encontrar valores unicos por columna por ejemplo para los paises, obtendremos los valores y los guardaremos en una lista desplegando paises en la posicion 150 a 160 (esto hecho arbitrariamente)

```python
name_country = data["name"].unique().tolist()
name_country[150:160]
```

Tambien podemos ordenar los paises de orden alfabetico

```python
sorted(name_country)[150:160]
```

Veamos la cantidad de datos por contienentes dados. Mostraremos los unicos que trae el dataframe

```python
names_continent = data["continent"].unique().tolist()
names_continent
```

Usando funciones como len() podemos determinar cuandos paises hay en los contienentes como: Africa,Asia y Europa

```python
name_continent_to = ["AF","AS","EU"]
for every_continent in name_continent_to:
    print("###################################################################")
    data_continent = data[data["continent"]==every_continent] 
    display(data_continent.head())
    print("Total de paises en el contienente",every_continent,"en el dataframe", len(data_continent))
```

Viendo un diagrama de caja de poblacion por pais-continente (solo para divertirnos con la información un poco)

```python
data.boxplot(column='area',by="continent",figsize=(10,7),fontsize=14)
```

```python
data["currency_name"].unique()
```

Podemos tambien filtrar información por mas criterios por ejemplo paises de Europa cuya moneda (currency_name) sea Euro o Dollar, ademas de tener un (area) area mayor a 1000

```python
data_Currency_area = data[((data["currency_name"]=="Euro") | (data["currency_name"]=="Dollar")) & (data["area"]>1000)]
display(data_Currency_area.head())
```

Podemos verificar los filtros hechos

```python
data_Currency_area["currency_name"].unique().tolist()
```

Usando un for para detectar areas menor o iguales a 1000

```python
for every_area in data_Currency_area["area"].unique().tolist():
    if every_area<=1000:
        print("Error")
        print(every_area)
    continue
```
